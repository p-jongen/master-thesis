```python
import socket

def list_ip_addresses(start_address, end_address):
    start = list(map(int, start_address.split(".")))
    end = list(map(int, end_address.split(".")))
    curr = start
    result = []
    while curr <= end:
        ip = ".".join(map(str, curr))
        result.append(ip)
        curr[3] += 1
        for i in (3, 2, 1):
            if curr[i] == 256:
                curr[i] = 0
                curr[i-1] += 1
    return result

start_address = "149.19.108.0"
end_address = "149.19.109.255"

ip_addresses = list_ip_addresses(start_address, end_address)

print("IP addresses in the given range:")
print("" + ",".join(ip_addresses) + "")
```
